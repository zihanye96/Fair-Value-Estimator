## Part C

Here, we will compute the mean order values for gender 0 and 1 and see if there's a significant difference.

```python
## split the dataset into two groups determined by gender

gender_zero = data[data["gender"] == 0]
gender_one = data[data["gender"] == 1]
```

```python
## get the group means, t-statistic, and p-value
mean_zero = gender_zero["value"].mean()
mean_one = gender_one["value"].mean()
t, p = stats.ttest_ind(gender_zero["value"], gender_one["value"], equal_var=False)

print("The mean order value for gender 0 is {} \nThe mean order value for gender 1 is {} \nT-statistic is {} with p-value of {}".format(mean_zero, mean_one, t, p))
```

From this dataset, it appears that for this store and this specific year, gender 0's mean order value is higher than that of gender 1 by approximately $13. 

To test for significance, I conducted a Welch's t-test, which assumes the two groups have unequal variances (Welch's t-test has similar power to a Student's t-test even when the two groups have equal variances), and the corresponding p-value is about .048. This means that if the two groups had the same mean order value, the chance of observing a difference as extreme or more than the one we observed is about 4.8\%. Since this is fairly unlikely, I believe that this difference is statistically significant and also practically significant, since $13 is not a negligible amount of money.

## Part D

Here, we will plot a confusion matrix for the gender predictions of customers in this dataset. I will assume that there's a unique gender prediction for each customer.

```python
## assume unique gender prediction for each customer
genders = data.groupby('customer_id').agg({'gender': np.mean, 
                                            'predicted_gender': np.mean}).sort_values(by = 'customer_id', ascending = True)
cm = confusion_matrix(genders["gender"],genders["predicted_gender"])
```

```python
print(cm)
```

Looking at the confusion matrix, it appears that there were 2189 correct predictions for gender 0 and 3512 correct predictions for gender 1, yielding a prediction accuracy of about 64.68%. The sensitivity is slightly better at 70.98%, but overall the confusion matrix indicates the predictions are not high quality, as there are a significant number misclassifications, especially for those who are actually gender 1, where more than 50% were misclassified.
